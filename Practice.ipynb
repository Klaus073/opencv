```python
import cv2 
import numpy as np
import matplotlib.pyplot as plt
```

```python
def negative_image_plot(old,new):
    plt.subplot(1,2,1)
    plt.title("old")
    plt.imshow(old)
    plt.show()
    plt.subplot(1,2,2)
    plt.title("new")
    plt.imshow(new)
    plt.show()
    
```

```python
old=cv2.imread("jett.jpg")
old=cv2.cvtColor(old,cv2.COLOR_BGR2RGB)
new=old.copy()
new=-1*new+255
```

```python
negative_image_plot(old,new)
```
